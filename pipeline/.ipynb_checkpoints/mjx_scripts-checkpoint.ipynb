# data preprocess v1 | discard

```python
from collections import defaultdict
import pandas as pd
import re
```

```python
file = 'uhgp.m8.f.drop.spread'
total_dict = defaultdict(lambda: defaultdict(int))
with open(file, 'r') as f:
    for row in f:
        name, genome, _ = row.strip().split()
        genome = genome.split('_')[0]
        total_dict[name][genome] += 1
```

```python
gene_info = pd.read_csv('./uhgp.m8.f', sep='\t', header=None, usecols=[0, 1], names=['gene','name'])
gene_info[['CF', 'CF_gene']] = gene_info['name'].apply(lambda x: pd.Series(x.split(',')))
genome_info = pd.read_csv('/data/database/uhgg/genomes-all_metadata.tsv', sep='\t', header=0)
genome_info['Genome'] = [re.sub(r'\.\d+$', '', x) for x in genome_info['Genome']]
```

```python
gene2CF = dict()
for x, y in zip(gene_info.gene, gene_info.CF):
    gene2CF[x] = y
```

```python
CF2genome2count = {}
for k, v in total_dict.items():
    CF = gene2CF[k]
    if CF not in CF2genome2count:
        CF2genome2count[CF] = {}
    for k2, v2 in v.items():
        if k2 not in CF2genome2count[CF]:
            CF2genome2count[CF][k2] = 0
        CF2genome2count[CF][k2] += v2
```

```python
# 所有检测到CF的genome
genome_list = set()
for x in total_dict.values():
    for y in x.keys():
        genome_list.add(y)

# 基因组和物种对应关系
genome2species = {}
for x, y in zip(genome_info['Genome'], genome_info['Species_rep']):
    if x in genome_list:
        genome2species[x] = y

# 物种对应基因组的列表
species2genome_list = {}
for x in genome_list:
    if genome2species[x] not in species2genome_list:
        species2genome_list[genome2species[x]] = []
    species2genome_list[genome2species[x]].append(x)
```

```python
# 累加同一物种物种所有基因组的CF
CF2speceis2count = {}
for k, v in CF2genome2count.items():
    if k not in CF2speceis2count:
        CF2speceis2count[k] = {}
    for k2, v2 in v.items():
        species = genome2species[k2]
        if species not in CF2speceis2count[k]:
            CF2speceis2count[k][species] = 0
        CF2speceis2count[k][species] += v2
```

```python
# 在物种内的基因组中流行率 > 50%
CF2speceis2binary = {}
for k, v in CF2genome2count.items():
    if k not in CF2speceis2binary:
        CF2speceis2binary[k] = {}

    for species in species2genome_list.keys():
        overlap = len(set(species2genome_list[species]).intersection(set(CF2genome2count[k].keys())))
        background = len(species2genome_list[species])
        if (overlap / background) >= 0.5:
           CF2speceis2binary[k][species] = 1
```

```python
data = pd.DataFrame(CF2speceis2binary).fillna(0)
```

```python
data.to_csv('./uhgp.m8.f.drop.spread.species.binary.tsv', sep='\t', index_label='name')
```

```python
from collections import defaultdict
import pandas as pd
import re
```

```python
gene_info = pd.read_csv('./uhgp.m8.f.drop', sep='\t', header=None, usecols=[0, 1], names=['gene','name'])
gene_info[['CF', 'CF_gene']] = gene_info['name'].apply(lambda x: pd.Series(x.split(',')))
gene2CF = dict()
for x, y in zip(gene_info.gene, gene_info.CF):
    gene2CF[x] = y
```

```python
gene2genome = {}
with open('uhgp.m8.f.drop.spread.genome_paste', 'r') as f:
    for row in f:
        gene, genome = row.strip().split()
        gene2genome[gene] = genome.split(';')
```

```python
CF2genome = {}
for k, v in gene2genome.items():
    CF = gene2CF[k]
    if CF not in CF2genome:
        CF2genome[CF] = set()
    for x in v:
        CF2genome[CF].add(x)
```

```python
# 所有检测到CF的genome
genome_list = []
for x in gene2genome.values():
    for y in x:
        genome_list.append(y)
genome_list = set(genome_list)
```

```python
# 基因组和物种对应关系
genome_info = pd.read_csv('/data/database/uhgg/genomes-all_metadata.tsv', sep='\t', header=0)
genome_info['Genome'] = [re.sub(r'\.\d+$', '', x) for x in genome_info['Genome']]
genome_info['Species_rep'] = [re.sub(r'\.\d+$', '', x) for x in genome_info['Species_rep']]
genome_info = genome_info[genome_info['Genome'].isin(genome_list)]

genome2species = {}
species2genome_list = {}
for x, y in zip(genome_info['Genome'], genome_info['Species_rep']):
    genome2species[x] = y
    if y not in species2genome_list:
        species2genome_list[y] = []
    species2genome_list[y].append(x)
```

```python
# 在物种内的基因组中流行率 > 50%
species_list = species2genome_list.keys()
CF2speceis2binary = {}

for k, v in CF2genome.items():
    if k not in CF2speceis2binary:
        CF2speceis2binary[k] = {}
    print(k)
    for species in species_list:
        overlap = len(set(species2genome_list[species]).intersection(v))
        background = len(species2genome_list[species])
        if (overlap / background) >= 0.5:
           CF2speceis2binary[k][species] = 1
```

```python
data = pd.DataFrame(CF2speceis2binary).fillna(0)
```

```python
data.to_csv('./uhgp.m8.f.drop.spread.species.binary', sep='\t', index_label='name')
```
